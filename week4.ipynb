```python
def sequences(filename):
    #Extract sequences and the identifiers in a single dictionary
    f=open(filename)
    s=dict()
    for line in f:
        line=line.rstrip()
        if line.startswith('>'):
            head=line.split()
            k=head[0]
            s[k]=" "
        else:
            s[k]=s[k]+line
    return s
```

```python
s=sequences("dna.example.fasta")
```

```python
for key,val in s.items():
    print(key,'\n',val)
```
